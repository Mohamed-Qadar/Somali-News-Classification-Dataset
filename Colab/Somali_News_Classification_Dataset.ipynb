# Dataset Description
This dataset contains Somali news articles collected from various online sources.
Each record represents a single news article and includes its textual content,
source information, and a categorical label indicating the news topic.

The dataset is primarily designed for:
- Text classification
- News topic categorization
- Natural Language Processing (NLP) experiments
- Low-resource language studies (Somali)

Expected columns:
- text    : News article content
- label   : Topic category of the article (e.g., Politics, World, Sports, Economy)
- source / website : Origin of the news article

```python
import pandas as pd
# Load the dataset

file_path = "/content/somali_news_PART1_FINAL_PW.csv"
df = pd.read_csv(file_path)

# Dataset size information

num_rows = df.shape[0]
print(f"Number of samples in the dataset: {num_rows}")

# Label analysis

if "label" in df.columns:
    num_unique_labels = df["label"].nunique()
    print(f"Number of unique labels: {num_unique_labels}")

    print("\nList of unique labels:")
    for label in df["label"].unique():
        print(f"- {label}")
else:
    print("No 'label' column found in the dataset.")

# Source analysis

if "source" in df.columns:
    print("\nUnique news sources:")
    for source in df["source"].unique():
        print(f"- {source}")

elif "website" in df.columns:
    print("\nUnique websites:")
    for website in df["website"].unique():
        print(f"- {website}")
else:
    print("\nNo 'source' or 'website' column found in the dataset.")

# Sample inspection

selected_labels = ["Politics", "World", "Sports", "Economy"]

for label in selected_labels:
    if "label" in df.columns:
        print(f"\n First 5 samples with label '{label}' \n")
        samples = df[df["label"] == label].head()
        display(samples)
```

```python
# -------------------------------------------------------------------
# Missing (NULL) value analysis
# This section checks whether the dataset contains missing values.
# It reports the number of null values per column and the total count.
# -------------------------------------------------------------------

# Check missing values per column
null_values_per_column = df.isnull().sum()

print("\nMissing values per column:")
print(null_values_per_column)

# Check if there is any missing value in the entire dataset
total_null_values = null_values_per_column.sum()
print(f"\nTotal number of missing values in the dataset: {total_null_values}")

# Boolean check for presence of missing values
if total_null_values == 0:
    print("The dataset does not contain any missing (NULL) values.")
else:
    print("The dataset contains missing (NULL) values.")
```

```python
columns_with_nulls = null_values_per_column[null_values_per_column > 0]
print("\nColumns containing missing values:")
print(columns_with_nulls)
```
